```python
import os

# Path to the volume directory containing the files
volume_path = '/Volumes/training_luc_chevre/bronze/'

# List all files in the volume directory
files = [f for f in os.listdir(volume_path) if os.path.isfile(os.path.join(volume_path, f))]

for file_name in files:
    file_path = os.path.join(volume_path, file_name)
    # Remove 'olist_' prefix if present and add '2' at the end before extension
    base_name = file_name
    if base_name.startswith('olist_'):
        base_name = base_name[len('olist_'):]
    table_name = os.path.splitext(base_name)[0] + '2'
    full_table_name = f'training_luc_chevre.bronze.{table_name}'
    
    # Infer file format (assuming CSV and Parquet are possible)
    if file_name.endswith('.csv'):
        df = spark.read.option("header", "true").csv(file_path)
    elif file_name.endswith('.parquet'):
        df = spark.read.parquet(file_path)
    else:
        continue  # Skip unsupported file types

    # Create table in the bronze schema
    df.writeTo(full_table_name).using("delta").createOrReplace()
```
